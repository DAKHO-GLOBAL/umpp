```python
from sqlalchemy import create_engine, MetaData
from sqlalchemy.exc import SQLAlchemyError

# Configuration de la base de données
db_config = {
    "host": "localhost",
    "user": "root",
    "password": "",
    "database": "pmu_ia",
    "port": "3306",
}

# Connexion à la base de données MySQL
engine = create_engine(f"mysql+mysqlconnector://{db_config['user']}:{db_config['password']}@{db_config['host']}:{db_config['port']}/{db_config['database']}")

# Créer les métadonnées sans 'bind'
metadata = MetaData()

# Fonction pour récupérer 2 lignes de chaque table
def get_two_rows_per_table(engine):
    connection = engine.connect()
    result = {}

    # Parcourir toutes les tables de la base de données
    metadata.reflect(bind=engine)  # Réfléchir les métadonnées en liant l'engine
    for table_name in metadata.tables:
        table = metadata.tables[table_name]
        try:
            query = table.select().limit(4)  # Sélectionner 2 lignes maximum
            rows = connection.execute(query).fetchall()
            
            # Ajouter les résultats dans le dictionnaire
            result[table_name] = rows
        except SQLAlchemyError as e:
            print(f"Erreur lors de la récupération des données pour la table {table_name}: {str(e)}")
    
    connection.close()
    return result

# Appel de la fonction pour récupérer les données
rows_per_table = get_two_rows_per_table(engine)

# Afficher les résultats
for table_name, rows in rows_per_table.items():
    print(f"Table: {table_name}")
    for row in rows:
        print(row)
    print('-' * 50)
```
